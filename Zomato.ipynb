```python
df_country=pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
final=pd.merge(df,df_country, on="Country Code",how="left")
```

```python
final.head(2)
```

```python
final.columns
```

```python
country_names=final.Country.value_counts().index
```

```python
country_values=final.Country.value_counts().values
```

```python
#pie Chart- top 3 that uses Zomato
plt.pie(country_values[:3],labels=country_names[:3],autopct="%.2f%%")
```

## Observation
India is mostly active in india
After that United States and United Kingdom and so on..

```python
final.columns
```

```python
ratings=final.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index().rename(columns={0:"Rating Count"})
ratings.head(5)
```

## Observation 
1. when rating is between 4.5 to 4.9 ---> Excelent
2. when rating is between 4.0 to 4.4 ---> very godd
3. when rating is between 3.5 to 3.9 --->Average
4. when rating is between `1.8 to 2.4 ---> Poor


```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",hue="Rating color",data=ratings,palette=["Cyan","Red","Orange","Yellow","Green","Blue"])
```

## Observation
1. Not rated count is very High
2. maximum number of rating are between 2.5 to 3.4


```python
#count Plot
sns.countplot(x='Rating color',data=ratings,palette=["Cyan","Red","Orange","Yellow","Green","Blue"])
```
